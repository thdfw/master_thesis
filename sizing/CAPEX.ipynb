```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
import casadi
import os
```

```python
# TES
A = [6, 9, 12]
C = [2416, 2857, 3297]
C = [x/365/20 for x in C]

C_polyfit = np.polyfit(A, C, 2)
sizes = list(range(2,30))
listt = [np.polyval(C_polyfit, x) for x in sizes]
plt.plot(sizes,listt,label='Approximated prices', linestyle='dashed')
plt.scatter(A,C,label='Reference prices')
plt.xlabel("Storage capacity (kWh)")
plt.ylabel("Price ($)")
plt.legend()
plt.savefig('capex_TES.png', dpi=200)
plt.show()
```
