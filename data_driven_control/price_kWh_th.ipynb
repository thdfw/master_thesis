```python
Q_max = pd.DataFrame(Q_max_list[1:], columns=Q_max_list[0]).T
Q_max.columns = Q_max.iloc[0]
Q_max = Q_max[1:]
print("Maximum Q_HP:")
display(Q_max)

Q_min = Q_max * 0.3
Q_min = Q_min.round(1)
print("Minimum Q_HP:")
display(Q_min)

COP = pd.DataFrame(COP_list[1:], columns=COP_list[0]).T
COP.columns = COP.iloc[0]
COP = COP[1:]
print("COP:")
display(COP)

W_max = Q_max / COP
W_max = W_max.round(1)
print("Maximum W_HP:")
display(W_max)

#print(Q_max[30][-7])
#print(list(Q_max[30]))
#print([Q_max[column][-7] for column in Q_max.columns])
```

```python
# Get the mass flow rate for a given LWT
def m_HP(T):
    if   T<=45: m = 34.5/60
    elif T<=55: m = 21.6/60
    elif T<=65: m = 17.3/60
    return round(m,2)

# Get the LWT options available at a given T_OA and T_HP_in
def get_LWT_options(T_OA, elec, T_HP_in):
    
    acceptable_LWT = []
    PRINT = False
    
    # Define the range of acceptable Q_HP for each LWT at the given T_OA
    LWT_options = [list[0] for list in Q_max_list][1:]
    Q_max_by_LWT = [Q_max[column][T_OA] for column in Q_max.columns]
    Q_min_by_LWT = [round(Q_min[column][T_OA]) for column in Q_min.columns]
    
    # Find the LWT that can work
    for i in range(len(LWT_options)):
        Q_HP = round(m_HP(LWT_options[i])*4187*(LWT_options[i]-T_HP_in),1)
        if Q_HP <= Q_max_by_LWT[i] and Q_HP >= Q_min_by_LWT[i]:
            if PRINT: print(f"[OK] LWT = {LWT_options[i]}°C, m_HP = {m_HP(LWT_options[i])} kg/s \
            gives Q_HP = {round(Q_HP/1000,1)} kW")
            acceptable_LWT.append(LWT_options[i])
        else:
            if PRINT: print(f"[--] LWT = {LWT_options[i]}°C, m_HP = {m_HP(LWT_options[i])} kg/s \
            gives Q_HP = {round(Q_HP/1000,1)} kW")
    
    # What are the corresponding prices
    #print(f"\nThe available LWTs are:")
    for LWT in acceptable_LWT:
        Q_HP = m_HP(LWT)*4187*(LWT-T_HP_in)/1000
        W_HP = Q_HP/COP[LWT][T_OA]
        if PRINT: print(f"{LWT}°C: Requires {round(Q_HP,1)} kW_th => {round(W_HP,1)} kW_elec => {round(W_HP*elec,2)}$ this hour => [{round(W_HP*elec/Q_HP,3)} $/kWh_th]")
        if PRINT: print(f"The acceptable range for this LWT is => [{round(Q_HP,1)}, {round(Q_max[LWT][T_OA]/1000)}]")

    Q_HP_min = round(Q_HP,3)
    Q_HP_max = round(Q_max[LWT][T_OA]/1000,3)
    cost_th = round(W_HP*elec/Q_HP,3)
    
    return Q_HP_min, Q_HP_max, cost_th
```

```python
# Set the (constant) incoming water temperature (43)
T_HP_in = 55 #°C
print(f"Assuming water incoming at {T_HP_in}°C")

T_OA_list = [2]*5 + [7]*15 + [2]*4 #°C
elec_list = [0.1, 0.15, 0.4, 0.4, 0.3] #$/kWh
elec_list = [6.36, 6.34, 6.34, 6.37, 6.41, 6.46, 6.95, 41.51, 41.16, 41.07, 41.06, 41.08, 7.16, 7.18, 7.18, 7.16, 41.2, 41.64, 41.43, 41.51, 6.84, 6.65, 6.46, 6.4]
load = [2, 10, 8, 20, 14] #kWh_th
load = [5.91, 5.77, 5.67, 5.77, 5.71, 6.06, 6.34, 6.34, 6.01, 5.77, 5.05, 5.05, 4.91, 4.91, 4.91, 4.91, 5.05, 5.1, 4.91, 4.91, 4.91, 4.91, 4.98, 4.91]

Q_HP_min_list = []
Q_HP_max_list = []
cost_th_list = []

for i in range(24):
    if PRINT: print(f"\n----- Hour {i+1} ({T_OA_list[i]}°C outside, {elec_list[i]*100} cts/kWh) ----")
    Q_HP_min, Q_HP_max, cost_th = get_LWT_options(T_OA_list[i], elec_list[i], T_HP_in)
    Q_HP_min_list.append(Q_HP_min)    
    Q_HP_max_list.append(Q_HP_max)
    cost_th_list.append(cost_th)
```

```python
import casadi, time
opti = casadi.Opti('conic')

# Variables
Q = opti.variable(24)
storage = opti.variable(25)
delta = opti.variable(24)
Q_times_delta = opti.variable(24)

# Delta is a discrete variable
discrete_var = [0]*24 + [0]*25 + [1]*24 + [0]*24

# Solver
opti.solver('gurobi', {'discrete':discrete_var, 'gurobi.OutputFlag':0})

# Constraints
for t in range(24):
    opti.subject_to(Q[t] <= Q_HP_max_list[t])    
    opti.subject_to(Q[t] >= Q_HP_min_list[t]*delta[t])
    opti.subject_to(storage[0] == 0)
    # Max storage capacity: opti.subject_to(storage[t] <= 30)
    opti.subject_to(storage[t+1] == Q_times_delta[t] - load[t] + storage[t])

    opti.subject_to(delta[t] >= 0)
    opti.subject_to(delta[t] <= 1)
    opti.subject_to(Q_times_delta[t] <= Q_HP_max_list[t]*delta[t])
    opti.subject_to(Q_times_delta[t] >= Q_HP_min_list[t]*delta[t])
    opti.subject_to(Q_times_delta[t] <= Q + Q_HP_min_list[t]*(delta[t]-1))
    opti.subject_to(Q_times_delta[t] >= Q + Q_HP_max_list[t]*(delta[t]-1))

for t in range(25):
    opti.subject_to(storage[t] > 0)

# Objective
obj = sum(Q_times_delta[t]*cost_th_list[t] for t in range(24))
opti.minimize(obj)

# Solve
sol = opti.solve()
Q_opt = sol.value(Q_times_delta)
stor_opt = sol.value(storage)

time.sleep(1)
print("")
print("Prices = ", cost_th_list)
print("Load = ", load)
print("Q_HP = ", [round(x,2) for x in Q_opt])
print("Storage = ", [round(x,2) for x in stor_opt])
```
