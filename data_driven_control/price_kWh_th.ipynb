```python
Q_max = pd.DataFrame(Q_max_list[1:], columns=Q_max_list[0]).T
Q_max.columns = Q_max.iloc[0]
Q_max = Q_max[1:]
print("Maximum Q_HP:")
display(Q_max)

Q_min = Q_max * 0.3
Q_min = Q_min.round(1)
print("Minimum Q_HP:")
display(Q_min)

COP = pd.DataFrame(COP_list[1:], columns=COP_list[0]).T
COP.columns = COP.iloc[0]
COP = COP[1:]
print("COP:")
display(COP)

W_max = Q_max / COP
W_max = W_max.round(1)
print("Maximum W_HP:")
display(W_max)
```

```python
# Get the mass flow rate for a given LWT
def m_HP(T):
    if   T<=45: m = 34.5/60
    elif T<=55: m = 21.6/60
    elif T<=65: m = 17.3/60
    return round(m,2)

# Get the LWT options available at a given T_OA and T_HP_in
def get_LWT_options(T_OA, elec, T_HP_in):
    
    acceptable_LWT = []
    
    # Define the range of acceptable Q_HP for each LWT at the given T_OA
    LWT_options = [list[0] for list in Q_max_list][1:]
    Q_max_by_LWT = [Q_max[column][T_OA] for column in Q_max.columns]
    Q_min_by_LWT = [round(Q_min[column][T_OA]) for column in Q_min.columns]
    
    # Find the LWT(s) that can work
    for i in range(len(LWT_options)):
        Q_HP = round(m_HP(LWT_options[i])*4187*(LWT_options[i]-T_HP_in),1)
        if Q_HP <= Q_max_by_LWT[i] and Q_HP >= Q_min_by_LWT[i]:
            if PRINT: print(f"[OK] LWT = {LWT_options[i]}°C, m_HP = {m_HP(LWT_options[i])} kg/s \
            gives Q_HP = {round(Q_HP/1000,1)} kW")
            acceptable_LWT.append(LWT_options[i])
        else:
            if PRINT: print(f"[--] LWT = {LWT_options[i]}°C, m_HP = {m_HP(LWT_options[i])} kg/s \
            gives Q_HP = {round(Q_HP/1000,1)} kW")
    
    # The corresponding prices and Q_HP ranges
    for LWT in acceptable_LWT:
        Q_HP = m_HP(LWT)*4187*(LWT-T_HP_in)/1000
        W_HP = Q_HP/COP[LWT][T_OA]
        if PRINT: print(f"\n{LWT}°C: Requires {round(Q_HP,1)} kW_th => {round(W_HP,1)} kW_elec => {round(W_HP*elec,2)}cts this hour => [{round(W_HP*elec/Q_HP,2)} cts/kWh_th]")
        if PRINT: print(f"The acceptable range for this LWT is => [{round(Q_HP,1)}, {round(Q_max[LWT][T_OA]/1000)}]")

    Q_HP_min = round(Q_HP,3)
    Q_HP_max = round(Q_max[LWT][T_OA]/1000,3)
    cost_th = round(W_HP*elec/Q_HP,3)
    
    return Q_HP_min, Q_HP_max, cost_th
```

```python
# The temperature of the water returning from the PCM to the HP
T_HP_in = 55 #°C
print(f"Assuming water incoming at {T_HP_in}°C")

# Forecasts (outside temp, elec prices, load)
T_OA_list = [2]*5 + [7]*15 + [2]*4 #°C
elec_list = [6.36, 6.34, 6.34, 6.37, 6.41, 6.46, 6.95, 41.51, 41.16, 41.07, 41.06, 41.08, 7.16, 7.18, 7.18, 7.16, 41.2, 41.64, 41.43, 41.51, 6.84, 6.65, 6.46, 6.4]
load = [5.91, 5.77, 5.67, 5.77, 5.71, 6.06, 6.34, 6.34, 6.01, 5.77, 5.05, 5.05, 4.91, 4.91, 4.91, 4.91, 5.05, 5.1, 4.91, 4.91, 4.91, 4.91, 4.98, 4.91]
load = [x/3 for x in load]

# Find the available range for Q_HP and the $/kWh_th for the available LWT
Q_HP_min_list = []
Q_HP_max_list = []
cost_th_list = []
for i in range(24):
    if PRINT: print(f"\n----- Hour {i+1} ({T_OA_list[i]}°C outside, {elec_list[i]} cts/kWh) ----")
    Q_HP_min, Q_HP_max, cost_th = get_LWT_options(T_OA_list[i], elec_list[i], T_HP_in)
    Q_HP_min_list.append(Q_HP_min)    
    Q_HP_max_list.append(Q_HP_max)
    cost_th_list.append(cost_th)
```

```python
import casadi, time
opti = casadi.Opti('conic')

# Horizon
N = 24

# Variables
Q_HP = opti.variable(1,N) # input
storage = opti.variable(1,N+1) # state
HP_on_off = opti.variable(1,N) # input
Q_HP_onoff = opti.variable(1,N) # input (derived)

# Delta is a discrete variable
discrete_var = [0]*N + [0]*(N+1) + [1]*N + [0]*N

# Solver
opti.solver('gurobi', {'discrete':discrete_var, 'gurobi.OutputFlag':0})

# Constraints on storage (state)
for t in range(N+1):

    # Initial and final storage
    opti.subject_to(storage[0] == 0)
    opti.subject_to(storage[N] == 5)
    
    # Minimum and maximum storage
    opti.subject_to(storage[t] >= 0)
    opti.subject_to(storage[t] <= 12)
    
    # System dynamics
    if t < N: opti.subject_to(storage[t+1] == Q_HP_onoff[t] - load[t] + storage[t])

# Constraints on Q_HP and HP_on_off (inputs)
for t in range(N):

    # Delta is binary
    opti.subject_to(HP_on_off[t] >= 0)
    opti.subject_to(HP_on_off[t] <= 1)

    # Q_HP is in a range
    opti.subject_to(Q_HP[t] <= Q_HP_max_list[t])    
    opti.subject_to(Q_HP[t] >= Q_HP_min_list[t]*HP_on_off[t])

    # Bilinear to linear
    opti.subject_to(Q_HP_onoff[t] <= Q_HP_max_list[t]*HP_on_off[t])
    opti.subject_to(Q_HP_onoff[t] >= Q_HP_min_list[t]*HP_on_off[t])
    opti.subject_to(Q_HP_onoff[t] <= Q_HP[t] + Q_HP_min_list[t]*(HP_on_off[t]-1))
    opti.subject_to(Q_HP_onoff[t] >= Q_HP[t] + Q_HP_max_list[t]*(HP_on_off[t]-1))

# Objective
obj = sum(Q_HP_onoff[t]*cost_th_list[t] for t in range(N))
opti.minimize(obj)

# Solve and get optimal values
sol = opti.solve()
Q_opt = sol.value(Q_HP_onoff)
stor_opt = sol.value(storage)
HP_on_off_opt = sol.value(HP_on_off)
obj_opt = round(sol.value(obj)/100,2)
```

```python
# Duplicate the last element of the hourly data for the plot
cost_th_list2 = cost_th_list + [cost_th_list[-1]]
Q_opt = [round(x,3) for x in Q_opt] + [Q_opt[-1]]
load2 = load + [load[-1]]

# Plot
fig, ax = plt.subplots(1,1, figsize=(13,5))
ax.step(range(25), Q_opt, where='post', label='Heat pump', alpha=0.4, color='blue')
ax.plot(range(25), stor_opt, label='Storage', alpha=0.6, color='orange')
ax.step(range(25), load2, where='post', label='Load', alpha=0.4, color='red')
ax.set_yticks(range(0,int(np.max(stor_opt)+5),4))
ax.set_xticks(range(25))
ax.set_ylabel("Energy [$kWh_{th}$]")
ax.set_ylim([0,15])
plt.legend()
plt.title(f"Cost: {obj_opt}$, Load: {round(sum(load2),1)} kWh_th \n=> {round(100*obj_opt/sum(load2),2)} cts/kWh_th")
ax2 = ax.twinx()
ax2.set_ylabel("Electricity price [$/kWh]")
ax2.step(range(25), cost_th_list2, where='post', label='Electricity price', color='black', alpha = 0.4)
plt.show()
```
