```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
import casadi
import os
```

```python
A = [24000, 36000, 48000]
A = [x/3412 for x in A]
C = [4333, 4844, 6713]
C_polyfit = np.polyfit(A, C, 3)
sizes = list(range(int(22000/3412),int(55000/3412)))
listt = [np.polyval(C_polyfit, x) for x in sizes]
plt.plot(sizes,listt,label='Approximated prices')
plt.scatter(A,C,label='Reference prices')
plt.xlabel("Heating capacity (kWh)")
plt.ylabel("Price ($)")
plt.legend()
plt.show()
```
