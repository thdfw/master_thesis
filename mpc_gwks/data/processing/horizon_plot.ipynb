```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
horizons = [0,2,4,8,16]
times = [0,9,15,41,75]
costs = [3.64,2.81, 2.45, 1.44, 1.36]
```

```python
fig, ax = plt.subplots(1,1, figsize=(12,4))
ax.plot(horizons,costs, label="Cost")
ax.scatter(horizons,costs)
ax.set_xlabel("Prediction horizon [hours]")
ax.set_ylabel("Cost of simulation [cts/kWh$_{th}$]")
ax.set_xticks(horizons)
ax2 = ax.twinx()
ax2.plot(horizons, times, color='tab:orange', label="Time")
ax2.scatter(horizons, times, color='tab:orange')
ax2.set_ylabel("Simulation time [minutes]")

lines1, labels1 = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax.legend(lines1 + lines2, labels1 + labels2)

plt.savefig(f"horizon_selection.pdf")
plt.show()
```
